### Dispo alimentaire

```python
# importation des données
dispoalim= pd.read_csv("DisponibiliteAlimentaire_2017.csv", sep=",")
dispoalim.head()
```

```python
#multplication des valeurs par milles
dispoalim["Valeur"]=dispoalim["Valeur"]*1000
dispoalim.head()
```

```python
# suppression de colonnes
dispoalim=dispoalim.drop(["Unité","Symbole","Description du Symbole","Code Domaine","Domaine"],axis=1)
dispoalim
```

```python
# changement du nom de colonne
dispoalim.rename(columns={"Valeur" : "Dispo_alimentaire"}, inplace=True)
dispoalim
```

```python
# vérification des nulls et du type colonne
dispoalim.info()
```

```python
# vérification des pays
dispoalim["Zone"].unique()
```

```python
#données du tableau avec le mot volaille dans les produits
dispoalim[dispoalim["Produit"].str.contains("Volaille")]
```

```python
# création du dataframe en incluant uniquement la viande de volaille
dispovolaille=dispoalim[dispoalim["Produit"]=="Viande de Volailles"]
dispovolaille
```

```python
# pivoting
alimvolaille=pd.pivot_table(dispovolaille, values="Dispo_alimentaire",index=["Zone","Code zone"],columns="Élément", aggfunc=np.sum).reset_index()
alimvolaille.head()
```

```python
# calcul pour en faire de nouvelles colonnes
alimvolaille["dispo_interieure"]=round(alimvolaille["Importations - Quantité"]+alimvolaille["Production"]+ alimvolaille["Variation de stock"]- alimvolaille["Exportations - Quantité"],1)
alimvolaille["auto_alimentation"]=round((alimvolaille["Production"]/alimvolaille["dispo_interieure"])*100,1)
alimvolaille["importation"]=round(alimvolaille["Importations - Quantité"]/alimvolaille["dispo_interieure"]*100,1)
alimvolaille
```

```python
#alimvolaille=alimvolaille.drop(["Aliments pour animaux","Autres utilisations (non alimentaire)", "Pertes","Semences","Traitement"],axis=1)
# choix des colonnes
alimvolaille=alimvolaille[["Code zone","Zone","dispo_interieure","auto_alimentation","importation"]]
alimvolaille
```

```python
# analyse de la colonne
#dispoalim[["Dispo_alimentaire"]].describe()
alimvolaille.describe()
```

```python
# pays unique
#dispoalim["Zone"].nunique()
alimvolaille["Zone"].nunique()
```

```python
# vérification des valeurs doubles
#dispoalim.duplicated().sum()
```

```python
# valeurs nulles
alimvolaille.isnull().sum()
```

```python
# top auto alimentation
auto_alim_top=alimvolaille.sort_values(by="auto_alimentation", ascending=False)
print(f"Les 15 pays pouvant ne compter que sur leur production pour se nourrir en volaille, sont ")
auto_alim_top.head(15)
```

```python
# flop auto alimentation
auto_alim_flop=alimvolaille.sort_values(by="auto_alimentation", ascending=True)
print(f"Les 15 pays ne pouvant pas s'auto alimenter en volaille, sont ")
auto_alim_flop.head(15)
```

```python
# top importation
import_top=alimvolaille.sort_values(by="importation", ascending=False)
print(f"Les 15 pays qui importent le plus de volailles, sont ")
import_top.head(15)
```

```python
# flop importation
import_flop=alimvolaille.sort_values(by="importation", ascending=True)
print(f"Les 15 pays qui importent le moins de volailles, sont ")
import_flop.head(15)
```

### Population

```python
# importation des données
population= pd.read_csv("Population_2000_2018.csv", sep=",")
population.head()
```

```python
# nombre de pays
population["Zone"].nunique()
```

```python
# pays unique
population["Zone"].unique()
```

```python
# pivoting pour avoir les années en colonne
pop=pd.pivot_table(population, values="Valeur",index=["Code zone","Zone"],columns="Année").reset_index()
pop.head()
```

```python
# vérification de la Chine pour vérifier si des incohérences existent
pop1=pop[pop["Zone"].str.contains("Chine")]
pop1
```

```python
# noms des colonnes
pop.columns
```

```python
# suppression de colonnes
pop=pop.drop(["Code zone",2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016],axis=1)
pop
```

```python
# changement de nomes de colonnes
pop.rename(columns={2017 : "pop_2017",2018:"pop_2018"}, inplace=True)
pop
```

```python
# multplication des valeurs par milles
pop[["pop_2017","pop_2018"]]=pop[["pop_2017","pop_2018"]]*1000
pop.head()
```

```python
# vérification du total de la population
pop["pop_2017"].sum()
```

```python
# calcul du taux de croissance
pop["taux de croissance"]=round(((pop["pop_2018"]-pop["pop_2017"])/pop["pop_2017"])*100,2)
pop
```

```python
# top croissance
poptop=pop.sort_values(by="taux de croissance", ascending=False)
print(f"Les 15 pays avec le meilleur taux de croissance, sont ")
poptop.head(15)
```

```python
# flop croissance
popflop=pop.sort_values(by="taux de croissance", ascending=True)
print(f"Les 15 pays avec le moins bon taux de croissance, sont ")
popflop.head(15)
```

```python
# informations du dataframe
pop.info()
```

```python
# valeurs nulles
popnul=pop[pop["pop_2017"].isnull() & pop["pop_2018"].isnull()]
popnul
```

```python
# dimensions du DF sans les deux pays 
pop = pop[~pop["pop_2017"].isnull()]
pop.shape
```

```python
# analyse de la population
pop.describe()
```

```python
# valeurs nulles
pop.isnull().sum()
```

```python
# vérification des valeurs doubles
pop.duplicated().sum()
```

```python
# valeurs doubles de la colonne
pop["Zone"].duplicated().sum()
```

## Jointure

```python
# jointure des dataframes 
dispopop=pd.merge(alimvolaille, 
pop, how="left", left_on=["Zone"], right_on=["Zone"])
dispopop
```

```python
# jointure des dataframes 
pibpop=pd.merge(dispopop, 
pib[["Zone","PIB"]], how="left", left_on=["Zone"], right_on=["Zone"])
pibpop
```
